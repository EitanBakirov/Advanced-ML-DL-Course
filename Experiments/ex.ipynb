```python
# Function to search for a track
def search_track(track_name):
    results = sp.search(q=track_name, type='track', limit=1)
    if results['tracks']['items']:
        return results['tracks']['items'][0]
    else:
        return None
```

```python
# Replace 'TRACK_NAME' with the name of the track you want to search for
track_name = '(The Angels Wanna Wear My) Red Shoes'
track_info = search_track(track_name)

# Print track info if found
if track_info:
    print("Track Name:", track_info['name'])
    print("Artist(s):", ", ".join([artist['name'] for artist in track_info['artists']]))
    print("Album:", track_info['album']['name'])
    print("Release Date:", track_info['album']['release_date'])
    print("Preview URL:", track_info['preview_url'])
    print("External URL:", track_info['external_urls']['spotify'])
else:
    print("Track not found.")
```

```python
# Function to get playlist URI by name
def get_playlist_uri_by_name(playlist_name):
    playlists = sp.search(q=playlist_name, type='playlist')
    if playlists['playlists']['items']:
        return playlists['playlists']['items'][0]['uri']
    else:
        print("Playlist not found")
        return None
```

```python
# Function to get songs in a playlist by playlist ID
def get_songs_in_playlist(playlist_id):
    # Retrieve tracks from the playlist
    playlist_tracks = sp.playlist_tracks(playlist_id)

    # Extract information for each track
    for idx, track in enumerate(playlist_tracks['items'], 1):
        track_name = track['track']['name']
        artists = ', '.join([artist['name'] for artist in track['track']['artists']])
        album_name = track['track']['album']['name']
        print(f"{idx}. {track_name} - {artists} - {album_name}")
```

```python
# Call the function to get songs in the playlist
get_songs_in_playlist("4BVbcWaTw03yqXQJxlijXv")
```

```python
# Function to get track info for all tracks in a playlist
def get_track_info_by_playlist(playlist_uri):
    playlist_tracks = sp.playlist_tracks(playlist_uri)["items"]
    track_info_list = []

    for track in playlist_tracks:
        track_info = track["track"]
        track_uri = track_info["uri"]
        artist_uri = track_info["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)
        artist_name = track_info["artists"][0]["name"]
        artist_popularity = artist_info["popularity"]
        artist_genres = artist_info["genres"]
        album_name = track_info["album"]["name"]
        track_name = track_info["name"]
        track_popularity = track_info["popularity"]

        # Append information for the current track
        track_info_list.append({
            "Track URI": track_uri,
            "Track Name": track_name,
            "Main Artist": artist_name,
            "Artist Popularity": artist_popularity,
            "Artist Genres": artist_genres,
            "Album": album_name,
            "Track Popularity": track_popularity
        })

    return track_info_list
```

```python
# Initialize Spotipy client
def get_track_info_by_track_name(track_name):
    # Search for the track by name
    results = sp.search(q=track_name, type='track', limit=1)

    # If track is found, extract information
    if results['tracks']['items']:
        track_info = results['tracks']['items'][0]
        track_uri = track_info["uri"]
        artist_uri = track_info["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)
        artist_name = track_info["artists"][0]["name"]
        artist_popularity = artist_info["popularity"]
        artist_genres = artist_info["genres"]
        album_name = track_info["album"]["name"]
        track_popularity = track_info["popularity"]
        # Get audio features for the track
        audio_features = sp.audio_features(track_uri)[0]
        danceability = audio_features["danceability"]
        mode = audio_features["mode"]
        tempo = audio_features["tempo"]

        # Return track information
        return {
            "Track URI": track_uri,
            "Track Name": track_name,
            "Main Artist": artist_name,
            "Artist Popularity": artist_popularity,
            "Artist Genres": artist_genres,
            "Album": album_name,
            "Track Popularity": track_popularity,
            "Danceability": danceability,
            "Mode": mode,
            "Tempo": tempo
        }
    else:
        print("Track not found.")
        return None
```

```python
# Example usage
track_name = "Alison"
track_info = get_track_info_by_track_name(track_name)
if track_info:
    print(track_info)
```
