```python
def extract_csv_from_rar(rar_content):
    # Create a RarFile object from the content
    with rarfile.RarFile(BytesIO(rar_content), 'r') as rar_file:
        # Get list of files in the RAR archive
        file_list = rar_file.namelist()

        # Initialize empty DataFrames
        df = pd.DataFrame()

        # Iterate through each file in the archive
        for file_name in file_list:
            # Check if the file is a CSV file
            if file_name.endswith('.csv'):
                print(file_name)
                # Extract the CSV file and read it into a DataFrame
                with rar_file.open(file_name) as csv_file:
                    df = pd.read_csv(csv_file)

    return df
```

```python
if song_data_rar_response.ok and count_data_rar_response.ok:
    # Extract content from the responses
    song_data_rar_content = song_data_rar_response.content
    count_data_rar_content = count_data_rar_response.content

    # Extract CSV files from RAR archives and convert to DataFrames
    song_df = extract_csv_from_rar(song_data_rar_content)
    count_df = extract_csv_from_rar(count_data_rar_content)
```

## Collaborative Filtering Model using Embeddings:

* Collaborative filtering is a technique commonly used in recommender systems.
* In the CF model, user and item embeddings are learned from the interactions (listen counts) between users and songs.
* The embeddings are then used to predict the listen counts for user-song pairs.
* The model computes the dot product of user and song embeddings to predict the score.
*This **bold text** model directly learns to predict the listen count for a user-song pair based on the interactions in the dataset.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Concatenate

# Take only the first 10,000 different users
user_subset = combined_df['user_id'].value_counts().index[:10000]
filtered_df = combined_df[combined_df['user_id'].isin(user_subset)]

# Convert user_id and song_id to categorical values
user_ids = filtered_df['user_id'].astype('category').cat.codes.values
song_ids = filtered_df['song_id'].astype('category').cat.codes.values
listen_counts = filtered_df['listen_count'].values

# Combine into a DataFrame
data = pd.DataFrame({'user_id': user_ids, 'song_id': song_ids, 'listen_count': listen_counts})

# Split the data into train, test, and validation sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
train_data, val_data = train_test_split(train_data, test_size=0.1, random_state=42)

# Define embedding sizes
embedding_size = 50

# Define input layers
user_input = Input(shape=(1,))
song_input = Input(shape=(1,))

# User embedding
user_embedding = Embedding(input_dim=len(data['user_id'].unique()),
                           output_dim=embedding_size)(user_input)
user_embedding = Flatten()(user_embedding)

# Song embedding
song_embedding = Embedding(input_dim=len(data['song_id'].unique()),
                           output_dim=embedding_size)(song_input)
song_embedding = Flatten()(song_embedding)

# Concatenate user and song embeddings
concat = Concatenate()([user_embedding, song_embedding])

# Fully connected layer
fc1 = Dense(128, activation='relu')(concat)

# Output layer
output = Dense(1)(fc1)

# Define the model
model = Model(inputs=[user_input, song_input], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit([train_data['user_id'], train_data['song_id']], train_data['listen_count'],
          validation_data=([val_data['user_id'], val_data['song_id']], val_data['listen_count']),
          epochs=5, batch_size=128)

# Evaluate the model on the test set
loss = model.evaluate([test_data['user_id'], test_data['song_id']], test_data['listen_count'])
print("Test Loss:", loss)

# Make predictions
predictions = model.predict([test_data['user_id'], test_data['song_id']])
```

## Neural Network-based Model:

* The neural network model is a more complex model compared to collaborative filtering.
* It also learns user and item embeddings, but it uses fully connected * layers to combine and process these embeddings.
* The model architecture consists of embedding layers, fully connected layers, and dropout layers for regularization.
* The fully connected layers allow for more flexibility and non-linear transformations of the embeddings.
* This model learns a more complex mapping from user-song pairs to listen counts, potentially capturing more intricate patterns in the data.
